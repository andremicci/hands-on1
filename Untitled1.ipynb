# Problema di approfondimento 1: Random walk

```python
import numpy as np
import matplotlib.pyplot as plt
import random as rnd
import math

```

```python
rnd.seed(121356) #inizializza la sequenza
rnd.random() #estrae numero tra 0 e 1

```

```python

pos=0
sqm=0
N=100
n=np.array([20,40,60])
plot=np.array
    
for i in range(0,N):
        
        d=2*np.sqrt(3)*rnd.random()-np.sqrt(3)
        pos=pos+d
        sqm=pos**2
        
        if i in n :
            plt.scatter(i,sqm)
            
        
            
            
        
        
        

```
